```python
import ccxt
import pandas as pd
import numpy as np
```

```python
# define exchange as coinbasepro

exchange = ccxt.coinbasepro()

# view markets (rough synonym for exchange rates between currencies)
markets = exchange.load_markets()
for market in markets:
    print(market)
```

```python
# create dataframe for ETH/USD, date is messed up

ETHUSD_ohlcv = pd.DataFrame(exchange.fetch_ohlcv('ETH/USD', timeframe = '1m'))
ETHUSD_ohlcv.columns = ['time', 'open', 'high', 'low', 'close', 'volume']
ETHUSD_ohlcv
```

```python
# Fix date

import datetime

date_fix = lambda x: datetime.datetime.fromtimestamp(int(x)/1000).strftime('%Y-%m-%d %H:%M:%S')
ETHUSD_ohlcv['time'] = ETHUSD_ohlcv['time'].apply(date_fix)
ETHUSD_ohlcv.set_index('time')
```
